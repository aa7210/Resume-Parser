```python
import json
from pdfminer.high_level import extract_text

def parse_resume(file_path):
    resume_content = extract_text(file_path)
    
    # Simple parsing - assuming sections are separated by newlines
    lines = resume_content.split('\n')
    resume_data = {
        "contact_info": "",
        "experience": [],
        "education": [],
        "skills": [],
        "summary": ""
    }

    section = ""
    for line in lines:
        line = line.strip()
        if not line:
            continue
        if "experience" in line.lower():
            section = "experience"
        elif "education" in line.lower():
            section = "education"
        elif "skills" in line.lower():
            section = "skills"
        elif "summary" in line.lower():
            section = "summary"
        elif "contact" in line.lower():
            section = "contact_info"
        else:
            if section == "experience" or section == "education":
                resume_data[section].append(line)
            elif section:
                resume_data[section] += line + " "

    return json.dumps(resume_data, indent=4)

# Example usage
file_path = "C:/Users/Aryan/OneDrive/Desktop/Resumes/Aryan_Resume.pdf"
parsed_resume_json = parse_resume(file_path)
print(parsed_resume_json)
```

```python
import fitz  # PyMuPDF
import json

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        text += page.get_text()
    return text

def parse_resume(pdf_path):
    resume_content = extract_text_from_pdf(pdf_path)
    
    # Simple parsing - assuming sections are separated by newlines
    lines = resume_content.split('\n')
    resume_data = {
        "contact_info": "",
        "experience": [],
        "education": [],
        "skills": [],
        "summary": ""
    }

    section = ""
    for line in lines:
        line = line.strip()
        if not line:
            continue
        if "experience" in line.lower():
            section = "experience"
        elif "education" in line.lower():
            section = "education"
        elif "skills" in line.lower():
            section = "skills"
        elif "summary" in line.lower():
            section = "summary"
        elif "contact" in line.lower():
            section = "contact_info"
        else:
            if section == "experience" or section == "education":
                resume_data[section].append(line)
            elif section:
                resume_data[section] += line + " "

    return resume_data

def save_to_json(data, output_file):
    with open(output_file, 'w') as json_file:
        json.dump(data, json_file, indent=4)

# Example usage
file_path = 'C:/Users/Aryan/OneDrive/Desktop/Resumes/Aryan_Resume.pdf'
parsed_resume = parse_resume(file_path)
save_to_json(parsed_resume, 'parsed_resume.json')
print("JSON file created successfully.")
```
